```python
import numpy as np
import matplotlib .pyplot as plt 
import pandas as pd
import seaborn as sns
```

```python
arr = np.random. randint (200,40000,100)
arr
```

```python
# lineplot 

arr.sort()
plt.plot(arr)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[19,20,37,40,57,60]
plt.plot(x,y,marker="s",ms=15,mfc="green",mec="red",color="black",lw=3,ls="--",alpha=0.5,label="lineplot")               
plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()

# ms -- marker size
# mfc -- marker face colour
# mec -- marker edge colour 
# color -- line colour 
# lw -- line weights
# ls -- line shape
# alpha -- transparancy
```

```python
#scattrplot
plt.scatter(x,y)
plt.show()
```
