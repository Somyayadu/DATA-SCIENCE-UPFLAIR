```python
import nltk
nltk.download('punkt_tab')
```

```python
corpus = "Bhagat Singh was one of the most influential revolutionaries of the Indian independence movement, remembered for his fearless spirit, deep patriotism, and intellectual strength. Born on 28th September 1907 in Banga, Punjab (now in Pakistan), Bhagat Singh grew up in a family that was deeply involved in the freedom struggle. Inspired by great freedom fighters like Kartar Singh Sarabha and by the oppressive policies of British colonialism, he was drawn into revolutionary politics at a very young age. He joined the Hindustan Socialist Republican Association (HSRA), which aimed to overthrow British rule through armed revolution and establish a society based on socialist principles. Bhagat Singh's bold actions, such as the assassination of British police officer John Saunders in 1928 to avenge the death of Lala Lajpat Rai, and the symbolic bombing of the Central Legislative Assembly in 1929, made him a national hero. Unlike many others, he did not run after these acts, but instead surrendered and used his trial and imprisonment as a platform to awaken the consciousness of the Indian masses. While in jail, he undertook hunger strikes and wrote extensively about his vision for a just and equal India, advocating for socialism and revolution through reason and education rather than blind violence. On 23rd March 1931, at the young age of 23, Bhagat Singh was executed along with his comrades Rajguru and Sukhdev."
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
doc = sent_tokenize(corpus)
for i in doc:
    print(i)
    print('\n')
    print(word_tokenize(i))
    print('\n')
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
